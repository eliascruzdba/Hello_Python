### Tratamento de valores nulos
### Ao realizar análises dependendo do case deletar uma linha inteira pode significar uma perda considerável na análise, por este motivo realizarmos o tratamento de Nulos através de diversas técnicas. Tudo irá depender do negócio, não existe uma solução, para isso depende de outros fatores do dado.

```python
#Projeção de colunas e falores nulos
df.isna().sum()
```

### Uma das técnicas utilizadas é a substituição do valor nulo pela média ou moda.

```python
df['km']
```

```python
#substituir pela moda ou média.
#Tratando dados Nulos e substituindo pela moda 
df['km'].fillna((df['km'].mode()), inplace=True)
```

### Realizar o merge entre bases é necessário para compor nossa análise. Para este exemplo vamos criar dois dataframes utilizando em comum o campo 'id_unico' para realizar o match entre as base.

### Merge entre duas bases

```python
df.columns
```

```python
df2 = df[['id_unico', 'dia_semana', 'horario', 'km', 'uf']]
```

```python
df3 = df[['id_unico', 'causa_acidente', 'tipo_acidente']]
```

```python
##Merge
df4 = pd.merge(df2,df3, on = 'id_unico', how = 'left')
```
